```python
ax = lcf[0].plot() 
ax.set_title("PDCSAP light curve of TIC 198382838")
```

```python
flat_lc = lcf[0].flatten(window_length=401)
flat_lc.plot();
```

```python
folded_lc = flat_lc.fold(period=9.868)
folded_lc.plot();
```

```python
binned_lc = folded_lc.bin(time_bin_size=0.01)
binned_lc.plot();
```
